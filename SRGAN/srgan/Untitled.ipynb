```python
import tensorflow as tf
import os
```

```python
gpus = tf.config.list_physical_devices('GPU')

if gpus:
  try:
    # Currently, memory growth needs to be the same across GPUs
    for gpu in gpus:
      tf.config.experimental.set_memory_growth(gpu, True)
    logical_gpus = tf.config.experimental.list_logical_devices('GPU')
    print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
  except RuntimeError as e:
    # Memory growth must be set before GPUs have been initialized
    print(e)
```

```python
filename = tf.data.Dataset.list_files("pokemon_ori\\*.png")
```

```python
str(next(iter(filename)).numpy().split()[0]).split("\\")[-1][:-1]
```

```python
def process_path(file_path):
    image = tf.io.read_file(file_path)
    image = tf.image.decode_png(image, channels=3)
    resize = tf.image.resize(image , size = (64,64))  
    
#     image = dataaugment(image)
#     resize = dataaugment(resize)
    
#     image = tf.expand_dims(image , axis = 0)
#     resize = tf.expand_dims(resize , axis = 0)
    
    return image , resize

def augment(image, resized):
    
    if tf.random.uniform((), minval=0, maxval=1) < 0.1:
        image = tf.tile(tf.image.rgb_to_grayscale(image), [1, 1, 3])
        resized = tf.tile(tf.image.rgb_to_grayscale(resized), [1, 1, 3])
        

    image = tf.image.random_brightness(image, max_delta=0.1)
    resized = tf.image.random_brightness(resized, max_delta=0.1)
    
    # a left upside down flipped is still a dog ;)
    image = tf.image.random_flip_left_right(image)  # 50%
    resizede = tf.image.random_flip_left_right(resized)
    # image = tf.image.random_flip_up_down(image) #%50%

    return image, resized
```

```python
dataset = filename.map(process_path)
dataset = dataset.map(augment).batch(8)
```
